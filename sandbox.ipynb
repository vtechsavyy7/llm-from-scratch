```python
import tiktoken
import numpy as np
```

```python
# Get the encoding for the 'gpt2' model
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
# Try tokenizing some text

text = "It's understandable that you're overwhelmed with the staggering amount of information available in supercalifragilisticexpialidocious."

tokens = tokenizer.encode(text)

print(" Tokens are: ", tokens)
print(" Number of tokens: ", len(tokens))
```

```python
# Get the vocabulary size of the tokenizer
vocab_size = tokenizer.n_vocab
print(" Vocabulary size: ", vocab_size)
```

```python
# Increment each token by 1 to create a new list of tokens
# incremented_tokens = [token + 1 for token in tokens]

# Decode the incremented tokens back to text
decoded_text = tokenizer.decode(tokens)

print(" Decoded text from incremented tokens: ", decoded_text)
```

```python
# Decode the original tokens individually to see the corresponding text for each token
decoded_tokens = [tokenizer.decode([token]) for token in tokens]

print(" Decoded tokens individually: ", decoded_tokens)
```
